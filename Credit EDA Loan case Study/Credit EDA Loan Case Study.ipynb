## Comparing on Credit_Range

```python
plt.figure(figsize=(15,5))

plt.subplot(1, 2, 1)
ax0 = defaulters["AMT_CREDIT_RANGE"].value_counts().sort_index().plot.bar()

plt.title('Distribution of AMT_CREDIT_RANGE for Defaulters',fontsize=15)
plt.xlabel('AMT_CREDIT_RANGE')
plt.ylabel('COUNT')

plt.subplot(1, 2, 2)
ax1 = non_defaulters["AMT_CREDIT_RANGE"].value_counts().sort_index().plot.bar()
plt.title(f'Distribution of AMT_CREDIT_RANGE for non-Defaulters',fontsize=15)
plt.xlabel("AMT_CREDIT_RANGE")
plt.ylabel('COUNT')  

total1 = float(len(defaulters))
total2 = float(len(non_defaulters))

# Annonating percentage for both chart.
for p in ax0.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total1)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax0.annotate(percentage, (x, y),ha='center')

for p in ax1.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total2)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax1.annotate(percentage, (x, y),ha='center')

plt.show()
```
